```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto=pd.read_csv('Automobile_data (1).csv')
auto.head()
```

```python
brandVSfuel=auto.groupby(["make","fuel-type"]).size().unstack(fill_value=0)
brandVSfuel
```

```python
brandVSfuel.plot(kind='bar')
plt.show
```

```python
brandVSdoors=auto.groupby(["make","num-of-doors"]).size().unstack(fill_value=0)
brandVSdoors
```

```python
brandVSdoors.plot(kind='bar',stacked=True)
plt.show
```

```python
#budget=5-10
#good=10-15
#best=15-20
#luxury=20 above
def car_type(p):
    if p=='?':
        return "price not available"
    x=int(p)    
    if x>=5000 and x<=10000:
        return "BUDGET CAR"
    elif x>=10000 and x<=15000:
        return "GOOD CAR"
    elif x>=15000 and x<=20000:
        return "Best car"
    elif x>=20000:
       return "LUXURY CAR"
auto['Car tpye']=auto['price'].apply(car_type)
auto
```

```python
auto.columns
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=auto["wheel-base"],linewidth=1.5,label="Wheel Base")
sns.lineplot(data=auto["engine-size"],linewidth=1.5,label="Engine Size")

plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")

plt.show()       
```

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#39FF14","axes.grid":False,"xtick.labelsize":10,"ytick.labelsize":10})

plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")
plt.xticks(rotation=40)
sns.lineplot(x=auto["engine-size"],y=auto["wheel-base"].index.values,color='r')
plt.show()
          
```
